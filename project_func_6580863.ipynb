```python
DROP TABLE IF EXISTS STUDENT_COURSE_MATERIAL CASCADE;
DROP TABLE IF EXISTS BOOKING CASCADE;
DROP TABLE IF EXISTS MATERIAL CASCADE;
DROP TABLE IF EXISTS ENROLLMENT CASCADE;
DROP TABLE IF EXISTS ROOM CASCADE;
DROP TABLE IF EXISTS COURSE CASCADE;
DROP TABLE IF EXISTS INSTRUCTOR CASCADE;
DROP TABLE IF EXISTS STUDENT CASCADE;
```

```python
# Student as the first entity
CREATE TABLE STUDENT (
    student_id SERIAL PRIMARY KEY,
    name TEXT NOT NULL,
    email TEXT UNIQUE NOT NULL,
    password_hash TEXT NOT NULL,
    usual_learning_days TEXT,
    usual_learning_time TEXT
);
```

```python
# Instructor as the second entity
CREATE TABLE INSTRUCTOR (
    instructor_id SERIAL PRIMARY KEY,
    name TEXT NOT NULL,
    email TEXT UNIQUE NOT NULL,
    instruments TEXT NOT NULL
);
```

```python
# Course as the third entity
CREATE TABLE COURSE (
    course_id SERIAL PRIMARY KEY,
    title TEXT NOT NULL,
    instrument TEXT NOT NULL,
    lesson_duration INT NOT NULL,
    total_hours INT NOT NULL,
    price NUMERIC(10,2) NOT NULL,
    difficulty TEXT
);
```

```python
# Room as the forth entity
CREATE TABLE ROOM (
    room_id SERIAL PRIMARY KEY,
    room_name TEXT NOT NULL,
    room_type TEXT NOT NULL,
    equipment TEXT
);
```

```python
# Enrollment as the fifth entity
CREATE TABLE ENROLLMENT (
    enrollment_id SERIAL PRIMARY KEY,
    student_id INT REFERENCES STUDENT(student_id),
    course_id INT REFERENCES COURSE(course_id),
    purchase_date DATE NOT NULL,
    hours_purchased INT NOT NULL,
    hours_used INT DEFAULT 0,
    expiration_date DATE NOT NULL
);
```

```python
# Booking as the sixth entity
CREATE TABLE BOOKING (
    booking_id SERIAL PRIMARY KEY,
    student_id INT REFERENCES STUDENT(student_id),
    instructor_id INT REFERENCES INSTRUCTOR(instructor_id),
    course_id INT REFERENCES COURSE(course_id),
    room_id INT REFERENCES ROOM(room_id),
    start_time TIMESTAMP NOT NULL,
    end_time TIMESTAMP NOT NULL,
    status TEXT DEFAULT 'scheduled',
    is_rescheduled BOOLEAN DEFAULT FALSE
);
```

```python
# Material the seventh entity
CREATE TABLE MATERIAL (
    material_id SERIAL PRIMARY KEY,
    title TEXT NOT NULL,
    type TEXT NOT NULL,
    file_url TEXT NOT NULL
);
```

```python
# STUDENT_COURSE_MATERIAL as the eight enity
CREATE TABLE STUDENT_COURSE_MATERIAL (
    id SERIAL PRIMARY KEY,
    student_id INT REFERENCES STUDENT(student_id),
    course_id INT REFERENCES COURSE(course_id),
    material_id INT REFERENCES MATERIAL(material_id),
    notes TEXT
);
```
