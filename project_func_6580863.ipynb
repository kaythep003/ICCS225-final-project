# **SQL Schema**

```python
%%sql
DROP TABLE IF EXISTS STUDENT_COURSE_MATERIAL CASCADE;
DROP TABLE IF EXISTS BOOKING CASCADE;
DROP TABLE IF EXISTS MATERIAL CASCADE;
DROP TABLE IF EXISTS ENROLLMENT CASCADE;
DROP TABLE IF EXISTS ROOM CASCADE;
DROP TABLE IF EXISTS COURSE CASCADE;
DROP TABLE IF EXISTS INSTRUCTOR CASCADE;
DROP TABLE IF EXISTS STUDENT CASCADE;
```

```python
# Student as the first entity
%%sql
CREATE TABLE Student (
    student_id SERIAL PRIMARY KEY,
    full_name VARCHAR(100) NOT NULL,
    email VARCHAR(120) UNIQUE NOT NULL,
    password_hash TEXT NOT NULL,
    paid_status BOOLEAN DEFAULT FALSE
);
```

```python
# Instructor as the second entity
%%sql
CREATE TABLE Instructor (
    instructor_id SERIAL PRIMARY KEY,
    full_name VARCHAR(100) NOT NULL,
    email VARCHAR(120) UNIQUE NOT NULL,
    instruments TEXT
);
```

```python
# Course as the third entity
%%sql
CREATE TABLE Course (
    course_id SERIAL PRIMARY KEY,
    course_name VARCHAR(120) NOT NULL,
    description TEXT,
    default_duration INTEGER CHECK (default_duration IN (30,60))
);
```

```python
# Room as the forth entity
%%sql
CREATE TABLE Room (
    room_id SERIAL PRIMARY KEY,
    room_name VARCHAR(10) UNIQUE NOT NULL
);
```

```python
# Enrollment (newly defined as Lesson) as the fifth entity
%%sql
CREATE TABLE Lesson (
    lesson_id SERIAL PRIMARY KEY,
    student_id INT REFERENCES Student(student_id) ON DELETE CASCADE,
    instructor_id INT REFERENCES Instructor(instructor_id) ON DELETE SET NULL,
    course_id INT REFERENCES Course(course_id) ON DELETE SET NULL,
    room_id INT REFERENCES Room(room_id),
    start_time TIMESTAMP NOT NULL,
    end_time TIMESTAMP NOT NULL,
    lesson_status VARCHAR(20) DEFAULT 'Scheduled'
);
```

```python
# Booking as the sixth entity
%%sql
CREATE TABLE BOOKING (
    booking_id SERIAL PRIMARY KEY,
    student_id INT REFERENCES STUDENT(student_id),
    instructor_id INT REFERENCES INSTRUCTOR(instructor_id),
    course_id INT REFERENCES COURSE(course_id),
    room_id INT REFERENCES ROOM(room_id),
    start_time TIMESTAMP NOT NULL,
    end_time TIMESTAMP NOT NULL,
    status TEXT DEFAULT 'scheduled',
    is_rescheduled BOOLEAN DEFAULT FALSE
);
```

```python
# Material the seventh entity
%%sql
CREATE TABLE MATERIAL (
    material_id SERIAL PRIMARY KEY,
    title TEXT NOT NULL,
    type TEXT NOT NULL,
    file_url TEXT NOT NULL
);
```

```python
# STUDENT_COURSE_MATERIAL as the eight enity
%%sql
CREATE TABLE STUDENT_COURSE_MATERIAL (
    id SERIAL PRIMARY KEY,
    student_id INT REFERENCES STUDENT(student_id),
    course_id INT REFERENCES COURSE(course_id),
    material_id INT REFERENCES MATERIAL(material_id),
    notes TEXT
);
```

```python
# prevent time conflict for both room booking and students' scheduling
%%sql
CREATE OR REPLACE FUNCTION check_time_conflict()
RETURNS TRIGGER AS $$
BEGIN
    IF EXISTS (
        SELECT 1 FROM Lesson
        WHERE room_id = NEW.room_id
        AND (
            (NEW.start_time < end_time AND NEW.end_time > start_time)
        )
    ) THEN
        RAISE EXCEPTION 'Time conflict in this room.';
    END IF;

    IF EXISTS (
        SELECT 1 FROM Lesson
        WHERE student_id = NEW.student_id
        AND (
            (NEW.start_time < end_time AND NEW.end_time > start_time)
        )
    ) THEN
        RAISE EXCEPTION 'Student already booked at this time.';
    END IF;

    RETURN NEW;
END;
$$ LANGUAGE plpgsql;
```

```python
%%sql
CREATE TRIGGER trg_time_conflict
BEFORE INSERT OR UPDATE ON Lesson
FOR EACH ROW EXECUTE FUNCTION check_time_conflict();
```

```python
# prevent rescheduling process that is less than 2 hrs before class time
%%sql
CREATE OR REPLACE FUNCTION prevent_last_minute_reschedule()
RETURNS TRIGGER AS $$
BEGIN
    IF OLD.start_time <> NEW.start_time THEN
        IF OLD.start_time - NOW() < INTERVAL '2 hours' THEN
            RAISE EXCEPTION 'Bookings cannot be changed within 2 hours.';
        END IF;
    END IF;
    RETURN NEW;
END;
$$ LANGUAGE plpgsql;
```

```python
%%sql
CREATE TRIGGER trg_no_reschedule
BEFORE UPDATE ON Lesson
FOR EACH ROW EXECUTE FUNCTION prevent_last_minute_reschedule();
```

# SQL functions according to the UI design

```python
# Login page -> verifying login
%%sql
CREATE OR REPLACE FUNCTION fn_login(p_email TEXT, p_password_hash TEXT)
RETURNS TABLE(student_id INT, full_name TEXT, paid BOOLEAN) AS $$
BEGIN
    RETURN QUERY
    SELECT student_id, full_name, paid_status
    FROM Student
    WHERE email = p_email AND password_hash = p_password_hash;
END;
$$ LANGUAGE plpgsql;
```

```python
# Dashboard page -> get information for upcoming class schedule + history for past classes taking
%%sql
CREATE OR REPLACE FUNCTION fn_dashboard(p_student INT)
RETURNS TABLE(
    lesson_id INT,
    start_time TIMESTAMP,
    end_time TIMESTAMP,
    lesson_status VARCHAR
)
AS $$
BEGIN
    RETURN QUERY
    SELECT lesson_id, start_time, end_time, lesson_status
    FROM Lesson
    WHERE student_id = p_student
    ORDER BY start_time;
END;
$$ LANGUAGE plpgsql;
```

```python
# booking page -> get available time slots
%%sql
CREATE OR REPLACE FUNCTION fn_available_times(
    p_room INT,
    p_date DATE
)
RETURNS TABLE(start_time TIMESTAMP, end_time TIMESTAMP) AS $$
BEGIN
    RETURN QUERY
    WITH hrs AS (
        SELECT generate_series(
            p_date + TIME '09:00',
            p_date + TIME '20:00',
            INTERVAL '30 minutes'
        ) AS t
    )
    SELECT t AS start_time, t + INTERVAL '30 min' AS end_time
    FROM hrs
    WHERE NOT EXISTS (
        SELECT 1 FROM Lesson
        WHERE room_id = p_room
        AND (t < end_time AND t + INTERVAL '30 min' > start_time)
    );
END;
$$ LANGUAGE plpgsql;
```

```python
# create booking inside booking page
%%sql
CREATE OR REPLACE FUNCTION fn_book(
    p_student INT,
    p_instructor INT,
    p_course INT,
    p_room INT,
    p_start TIMESTAMP,
    p_duration INT
)
RETURNS INT AS $$
DECLARE new_id INT;
BEGIN
    INSERT INTO Lesson (
        student_id, instructor_id, course_id, room_id,
        start_time, end_time
    )
    VALUES (
        p_student, p_instructor, p_course, p_room,
        p_start, p_start + (p_duration || ' minutes')::INTERVAL
    )
    RETURNING lesson_id INTO new_id;

    RETURN new_id;
END;
$$ LANGUAGE plpgsql;
```

```python
# reschedule function inside the booking page
%%sql
CREATE OR REPLACE FUNCTION fn_reschedule(
    p_lesson INT,
    p_new_start TIMESTAMP
)
RETURNS VOID AS $$
DECLARE duration INTERVAL;
BEGIN
    SELECT end_time - start_time INTO duration
    FROM Lesson WHERE lesson_id = p_lesson;

    UPDATE Lesson
    SET start_time = p_new_start,
        end_time = p_new_start + duration
    WHERE lesson_id = p_lesson;
END;
$$ LANGUAGE plpgsql;
```
